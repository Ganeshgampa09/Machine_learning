```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

```python
dataset = pd.read_csv('diabetes.csv')
```

```python
dataset.head()
```

```python
dataset.shape
```

```python
dataset.groupby('Outcome').mean()
```

```python
X = dataset.drop('Outcome',axis=1)
Y = dataset['Outcome']
```

```python
print(X)
```

```python
print(Y)
```

Splitting the data into Training data and Testing data

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=2)
```

```python
print(X_train)
```

```python
print(X.shape ,X_train.shape , X_test.shape)
```

Training the modal:-
support vector machine has two types:-
1)SVC :- support vector classifier
2)SVR :- support vector regressor
SVC is used for classification(categorical)
SVR is used for regression (Numerical)

Training the modal

```python
classifier = svm.SVC(kernel='linear')
```

```python
classifier.fit(X_train , Y_train)
```

```python
#accuracy score on the training data
X_train_prediction = classifier.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction ,Y_train)
```

```python
print(training_data_accuracy)
```

```python
X_test_prediction = classifier.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction,Y_test)
```

```python
print(test_data_accuracy)
```

Making a predictive system

```python
input_data = (1,115,70,30,96,34.6,0.529,32)

input_data_as_numpyarray = np.asarray(input_data)
```

```python
print(input_data_as_numpyarray)
```

```python
#we need to reshape the data because our modal is trained on 768 input now we just want to predict only for one instance.So we need to reshape the data.
input_data_reshaped = input_data_as_numpyarray.reshape(1,-1)
```

```python
print(input_data_reshaped)
```

```python
prediction = classifier.predict(input_data_reshaped)
print(prediction)

if prediction[0] == 0:
  print("The person don't have diabetic")
else:
  print("The person having diabetic")
```
