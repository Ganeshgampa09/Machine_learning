Imbalanced Dataset:
A dataset with an unequal class distribution

```python
import numpy as np
import pandas as pd
```

```python
credit_data = pd.read_csv('credit_data.csv')
```

```python
credit_data.shape
```

```python
credit_data.head()
```

```python
credit_data['Class'].value_counts()
```

This is highly imbalanced dataset

0 represents -- Legit transactions


1 -- Fraudulent Transactions

```python
legit = credit_data[credit_data.Class == 0]
fraud = credit_data[credit_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
legit_sample = legit.sample(n=156)
```

```python
print(legit_sample.shape)
```

```python
#concatenate the Two dataframes

new_dataset = pd.concat([legit_sample,fraud], axis=0)
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

In the credit_data.csv dataset there is a column called class that specifies whether the transaction is legit(geniune) or fraud. suppose let's say there is 10000 rows in the dataset and 9600 are legit transactions and 400 are fraud transaction.Then this is called as imbalanced dataset and also modal cannot train correctly.
