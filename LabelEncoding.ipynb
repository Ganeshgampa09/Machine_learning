Label_Encoding :- The process converting the labels into numerics.

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
data = pd.read_csv('/content/breast_cancer_data.csv')
```

```python
print(data)
```

```python
data.head()
```

```python
data['diagnosis'].value_counts()
```

```python
label_encoder = LabelEncoder()
```

```python
label = label_encoder.fit_transform(data.diagnosis)
```

```python
data['target'] = label
```

```python
data.tail()
```

```python
data['diagnosis'].value_counts()
```
